```python
# identify null values for country
# bot.country[bot.country.isnull()]

# replace null values for country with USA
bot.country.replace([np.nan], ['usa'], inplace = True)

# convert country text into feature vectors and 
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(decode_error = 'ignore')
vect.fit(bot.country)
vect.get_feature_names()

# transform country into a 'document-term matrix'
country_dtm = vect.transform(bot.country)
country_dtm.toarray() 
country_data = pd.DataFrame(country_dtm.toarray(), columns=vect.get_feature_names())
country_data.sum()
```

```python
# dummy_df=pd.get_dummies(bot['country'], prefix='country')
# dummy_df

bot_dict = {'genre':'nominal', 'director':'nominal', 'writer':'nominal', 'cast':'nominal', 'language':'nominal', 'country':'nominal'}

def dummy_variables(data, data_type_dict):
    #Loop over nominal variables.
    for variable in filter(lambda x: data_type_dict[x]=='nominal',
                           data_type_dict.keys()):
 
        #First we create the columns with dummy variables.
        #Note that the argument 'prefix' means the column names will be
        #prefix_value for each unique value in the original column, so
        #we set the prefix to be the name of the original variable.
        dummy_df=pd.get_dummies(data[variable], prefix=variable)
 
        #Remove old variable from dictionary.
        data_type_dict.pop(variable)
 
        #Add new dummy variables to dictionary.
        for dummy_variable in dummy_df.columns:
            data_type_dict[dummy_variable] = 'nominal'
 
        #Add dummy variables to main df.
        data=data.drop(variable, axis=1)
        data=data.join(dummy_df)
 
    return [data, data_type_dict]

new_df = pd.DataFrame(dummy_variables(bot, bot_dict)[0])
```

```python
genre_list = set(bot.genre.values)
feature_prefix = "genre_"
```

```python
def combine_features(row):
    
    new_column = []
    
    for key in genre_list:
        new_column.append(row[feature_prefix + key])
    
    return new_column
    
new_df.head(1).apply(combine_features)
```
